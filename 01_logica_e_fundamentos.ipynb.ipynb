```python
#conjuntos ou sets (coleções não ordenadas de ítens únicos e mutáveis) são excelentes para remover duplicatas e aplicar operações matemáticas
#dicionário: abre chaves e passa os pares chave 
numeros = {1,2,2,3,4,5,6,6}
```

```python
type(numeros)
```

```python
#observe que quando se dá o comando print no conjunto, os repetidos não entram
print(f'os números do conjunto são {numeros}')
```

```python
#adiciona um novo elemento no conjunto
numeros.add(7)
```

```python
#não aceita diversos elementos
numeros.add(8,9,10)
```

```python
print(f'o seu conjunto contém: {numeros}')
```

```python
conjunto_a = {1,2,3,4,5,5,6}
conjunto_b = {6,7,7,8,9,10}
```

```python

uniao = conjunto_a.union(conjunto_b)
print(f'A união de A e B: {uniao}')
```

```python
interseccao = conjunto_a.intersection(conjunto_b)
print(f'interseccao de A e B: {interseccao}')
```

```python
#convertendo inteiro para float
#ficar atento, número float é com ponto e não com vírgula
numero_decimal_em_texto = "35.29"
numero_float = float(numero_decimal_em_texto)
print(f'float \ {numero_decimal_em_texto}\' para float: {numero_float}, tipo: {type(numero_float)}')
```

```python
#convertendo estruturas de dados
#convertendo lista para conjunto, vai remover as duplicadas, gravando em outro objeto
lista_com_duplicadas= [1,2,3,3,3,4,5,6]
conjunto_unico= set(lista_com_duplicadas)
```

```python
lista_sem_duplicadas= list(conjunto_unico)
```

```python
print(f"lista original: {lista_com_duplicadas}")
```

```python
print(f'conversão para conjunto (remove duplicadas): {conjunto_unico}')
```

```python
print(f'lista sem duplicadas: {lista_sem_duplicadas}')
```
